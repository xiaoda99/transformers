# 模型之间差距

```python
#gptj - gptnew
models = [('task_accuary_gptj', 'task_accuary_gpt_neox'),('task_accuary_gptj', 'task_accuary_llama_7B_space_0329')]
for model1, model2 in models:
    dictnew = {}
    for key in keys:
        dictnew[key] = {'texts': dicts[model1][key]['texts'], \
                        f'{model1}' : dicts[model1][key]['mean_acc'], \
                        f'{model2}' : dicts[model2][key]['mean_acc'], \
                        'acc_cha' : dicts[model1][key]['mean_acc'] - dicts[model2][key]['mean_acc']}
    listss = sorted(dictnew.items(), key = lambda x: x[1]['acc_cha'], reverse = True)
#     print(listss[0])
#     for i in range(3):
#         print(listss[i])
#     for i in range(-3, 0):
#         print(listss[i])
    x_y = [(line[0],line[1]['acc_cha'] )for line in listss]
    x = [i[0] for i in x_y]
    y = [i[1] for i in x_y]
    plt.plot(range(len(y)),y)
    plt.xlabel(f'{model1} - {model2}')
    plt.ylabel('mean_acc')
    plt.show()
```
